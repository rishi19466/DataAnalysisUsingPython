### 2. Data Loading and Description!


- The dataset consists of the information about people Suffering from heart disease. Various variables present in the dataset includes data of age, sex etc. 
- The dataset comprises of __303 observations of 14 columns__. Below is a table showing names of all the columns and their description.

| Column Name   | Description                                                        |
| ------------- |:-------------                                                     :|
| age           | Age of person                                                      |
| sex           | Sex of person male or female                                       |
| cp            | Type of chest pain                                                 |
| trestbps      | resting blood pressure (in mm Hg on admission to the hospital      |
| chol          | serum cholestoral in mg/dl                                         |
| fbs           | (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)            |
| restecg       | resting electrocardiographic results                               |
| thalach       | maximum heart rate achieved                                        |
| exang         | exercise induced angina (1 = yes; 0 = no)                          |
| oldpeak       | ST depression induced by exercise relative to rest                 |
| slope         | the slope of the peak exercise ST segment                          |
| ca            |number of major vessels (0-3) colored by flourosopy                 |
| thal          |3 = normal; 6 = fixed defect; 7 = reversable defect                 |
| target        |0 or 1

### 4. Question

###  4.1 After what age people have a chance of getting a heart disease ?

```python
sns.boxplot(y=data['age'])
plt.title('Box plot of Age', size = 30, color = 'r')
```

#### From box plot we can observe that most of the observation are between age 47 to 61 which means mostly patient get heart disease after age of 45

### 4.2 Relatiob between age and trestbps

```python
# scatter plot 
sns.jointplot('age','trestbps',data)
```

```python
# Now lets get deeper 
df1 = data[(data.trestbps<160) & (data.trestbps>100)& (data.age>40)& (data.age<70)]
```

```python
sns.jointplot(x= 'age', y = 'trestbps',data = data)
```

```python
sns.jointplot('age','trestbps',df1, kind = 'hex')
```

### 4.3 Relation between age and chol

```python
sns.jointplot('age','chol',data)
```

```python
sns.jointplot('age','thalach',data)
```

```python
df2 = data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']]
sns.pairplot(df2)
```

```python
plt.figure(figsize = (15,12))
sns.heatmap(data.corr())
plt.title("corelation graph")
```

### 4.4 number of male and female with heart disease

```python
sns.countplot(x='target', data = data, hue = 'sex')
plt.title('Number of Male & Female with heart disease')
plt.show()
```

```python
#Chest Pain

sns.countplot(x='target', hue= 'cp',data = data)
plt.title('Chest Pain Type')
plt.show()
```

### There are 4 types of chest pain(0,1,2,3) out of which cp = 2 is the most common type of chest pain in case of people who have heart disease.

### Those who do not have a heart disease experience cp = 0 type of chest pain.

### 4.5 Have heart disease or not 

```python
sns.countplot(x ='target',data = data)
plt.title('Have heart disease or not')
```
