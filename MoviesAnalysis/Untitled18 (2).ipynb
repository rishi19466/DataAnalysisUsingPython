```python
movies_data['Year'].value_counts().sort_index().plot.bar()
```

```python
plt.show()
```

```python
Average_Rating = movies_data.groupby('Year').Rating.mean()    # Calculating avg rating of each year
```

```python
Average_Rating
```

```python
Average_Rating.plot.bar()
```

```python
plt.show()
```

   Average ratings were almost same during period 2008-14, but have started to decline in last two years.

```python
# visualize the relationship between Rating and duration
movies_data.sample(100).plot.scatter(x='Rating',y='Runtime')
```

 Most of the movies have run time between 100-140 min. Mostly, longest duration movies have higher ratings.


```python
visual = sns.lmplot(data=movies_data.sample(100), x='Runtime', y='Revenue', \
                    fit_reg=False,hue='Year')
```

```python
print('Revenue in millions')
print('maximum Revenue', movies_data['Revenue'].max())
print('minimum Revenue', movies_data['Revenue'].min())
```

```python
j = sns.jointplot(data=movies_data, x='Rating', y='Metascore', kind = 'hex')
```
